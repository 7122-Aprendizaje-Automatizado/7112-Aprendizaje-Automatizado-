```python
from numpy import random
from scipy import stats as st
from matplotlib import pyplot as plt

import numpy as np
import matplotlib

#a continuacin vamos a simular el lanzamiento de dos datos para sumar la cara superior
#para hacer esto debemos simular el experimento una gran cantidad de veces
#y guardarlo.

#lista para guardar suma
H = []

#simulacin
for i in range (1, 20001):
  tirada = random.choice([1,2,3,4,5,6])
  tirada2 = random.choice([1,2,3,4,5,6])
  #VA
  suma = tirada + tirada2
  H.append(suma)

H = np.array(H)
##print(H)

print(np.mean(H))
print(st.mode(H))

fig = plt.figure(figsize=(8,10))
plt.plot(H[:100])
```
